```python
m = {'a':1,'b':2}
```

```python
for i in m.keys() :
    print(i)
```

```python
for i in m.values() :
    print(i)
```

```python
for (k,v) in m.items() :
    print(k,v)
```

## 순환문 제어하기

- 반복되는 것을 중간에 종료하거나 처리를 다음으로 넘기헤 할 수 있음


###  순환문 강제 종료 

```python
for i in range(100000):
    print('loop')
    break
```

### 순환문 강제 점프

```python
for i in range(10):
    if i % 2 == 0 :
        continue
    print('loop')
    
```

### 순환문 실행한 후에 특정 로직을  반드시 처리 하기

```python
for i in range(1):
    print('for')
else:
    print('for else')
```

###  강제 종료하면  else 문은 처리하지 않음

```python
for i in range(1000):
    print('for')
    break
else:
    print('else')
```

##   while 문 

- 특정 조건식이 만족하는 경우에 순환을 처리


## while 순환문은 조건식에 따라 순환여부를 결정

### 순환문 강제 종료 

```python
while True:
    print("loop")
    if True:
        break
```

### 순환문 중간에 점프하기 :  continue

```python
n = 0
while n < 2:
    print('start')
    n += 1
    continue
    print('end')
```

### 반드시 처리 하기

```python
n = 0
```

```python

while n < 1:
    print("While")
    n += 1
else:
    print("While ELSE")
```

## 스위치 케이스를 딕셔너리로 처리하기

```python
import operator as op
```

```python
for i in dir(op) :
    print(i, end=", ") if not i.startswith('_') else _
```

```python
def fop(x,y,ops): 
    return {'+': op.add, '-': op.sub, '*':op.mul}.get(ops, op.add)(x,y)
```

```python
fop(10,10,'+'), fop(10,10,"*")
```

# 3. 예외문 

- 모든 코드는 실행 중에 에러나 예외가 발생할 때도 특정 코드를 처리할 필요가 있다.
- 이럴 경우 예외문을 사용해서 정상적인 처리가 가능하게 코딩을 처리해야 한다 



### try

    예외처리를 할 때 사용하는 문장
    이때도 예외가 발생하지 않으면 else 문을 수행

### 예외가 있는 경우 else 문 처리하지 않음 

```python
try:
    print("try")
    assert random.randint(0, 1)
    # A Error
    # B Error
except:
    print("Exception")
else:
    print('else')
```

### 예외가 없는 경우 else 문 처리

```python
try:
    print("try")
except:
    print("Exception")
else:
    print('else')
```

## finally
      예외가 없으면 반드시 처리 

```python
try:
    print("Try")
except:
    print("Except")
else:
    print('Else')
finally:
    print("finally")
```

### 예외를 발생시키면 현재 프로세스가 죽는다.

```python
x = int(input('나눌 숫자를 입력하세요: '))
y = 10 / x
```

### 특정 로직에서 예외를 발생해도 처리가 가능하도록 정의.

```python
try:
    x = int(input('나눌 숫자를 입력하세요: '))
    y = 10 / x
    print(y)
except ZeroDivisionError as e :
    print('ZeroDivisionError 예외가 발생했습니다.', e)
except Exception as e :    # 예외가 발생했을 때 실행됨
    print('Exception 예외가 발생했습니다.', e)
```

# 4. 컴프리헨션 이해하기

- 시퀀스 내의 원소를 특정 값으로 변경하거나 특정 조건에 따른 값을 선택할 때 작성한다.
- 특정 시퀀스의 원소들을 로직으로 표현할 수 있다.

- 원소가 많은 경우 로직으로 축약해서 보여준다.



### 리스트를 축약해서 표현하기 

```python
l = [ x for x in [1,2,3,4]]
```

```python
l
```

### map 으로 처리하기

- map은 원소를 특정 함수로 처리해서 변경한다.
- 익명함수는 lambda 매개변수 : 표현식으로 작성한다.

```python
map(lambda x : x, l)
```

```python
[*map(lambda x : x, l)]
```

### 리스트 내부 값을 변형하기 

```python
l = [ x*2 for x in [1,2,3,4]]
```

```python
l
```

### 맵으로 처리

```python
[*map(lambda x : 2*x, [1,2,3,4])]
```

###  집합을  축약해서 표현하기

```python
s = { x for x in [3,4,5,6]}
```

```python
s
```

### 딕셔너리를 축약해서 표현하기

```python
m = { k:v for k,v in [('a',3),('b',4)]}
```

```python
m
```

### 컴퓨리헨션에 가드 주기 

- 특정 조건을 제거하고 처리한다

```python
l2 = [ x for x in [1,2,3,4] if x % 2 == 0]
```

```python
l2
```

### filter로 처리

- 특정 원소를 참값만 추출하기

```python
[*filter(lambda x : x%2 == 0, [1,2,3,4])]
```
