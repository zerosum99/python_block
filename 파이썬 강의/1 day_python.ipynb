###  Sequence 타입이나 map/set 타입은 순환문을 통해 원소들을 가져올 수 있음

```python
for i in [1,2,3,4] :
    print(i)
```

#### map 타입은 순환문을 처리할 때는 키 값만 가져옴

## 순환문 제어하기

###  순환문 강제 종료 

```python
for i in range(100000):
    print('loop')
    break
```

### 순환문 강제 점프

```python
for i in range(10):
    if i % 2 == 0 :
        continue
    print('loop')
    
```

### 반드시 처리 하기

```python
for i in range(1):
    print('for')
else:
    print('for else')
```

#### 강제 종료하면 반드시 처리하지 않음

```python
for i in range(1000):
    print('for')
    break
else:
    print('else')
```

## 2-2   while 문 

## while 순환문은 조건식에 따라 순환여부를 결정

### 순환문 강제 종료 

```python
while True:
    print("loop")
    if True:
        break
```

### 순환문 중간에 점프하기 :  continue

```python
n = 0
while n < 2:
    print('start')
    n += 1
    continue
    print('end')
```

### 반드시 처리 하기

```python
n = 0
```

```python

while n < 1:
    print("While")
    n += 1
else:
    print("While ELSE")
```

# 3. 조건문

### 순환문이나 조건문에 else 처리

     특정 조건에 따라 else 문을 수행한다. 

## 3.1. 단순 조건문  if 

     모든 조건이 불일치시 else 문 수행 

```python
import random

if random.randint(0, 1):
    print('True')
else:
    print("False")
```

## 3.2. 복합 조건문  elif 


```python
a = 5
```

```python
if a > 10 :
    print('aaa')
elif a <= 5 :
    print("bbb")
else :
    print('else')
```

## 3.3. 삼항연산자

```python
print('a') if a > 5 else print("xxx")
```

# 4. 예외문 

### try

    예외처리를 할 때 사용하는 문장
    이때도 예외가 발생하지 않으면 else 문을 수행

```python
try:
    print("try")
    assert random.randint(0, 1)
    # A Error
    # B Error
except:
    print("Exception")
else:
    print('else')
```

```python
try:
    print("try")
except:
    print("Exception")
else:
    print('else')
```

## finally
      예외가 없으면 반드시 처리 

```python
try:
    print("Try")
except:
    print("Except")
else:
    print('Else')
finally:
    print("finally")
```
