## 함수의 합성

한 함수의 공역이 다른 함수의 정의역과 일치하는 경우 하나의 함수로 만드는 연산

`f(g(x)) == f*g(x)`

````python
def add_3(value): # x + 3
    return value + 3

def mul_2(value): # 2x
    return value * 2
````

````python
# f(g(x))
result1 = mul_2(3)
result2 = add_3(result1)
result2
````

### 함수의 합성 compose

````python
def compose(f, g):
    return lambda x : f(g(x))
````

````python
new_func = compose(add_3, mul_2)
````

````python
new_func(3)
````

````python
from functools import reduce


def compose2(*func):
    # func = (f, g, h)
    # func = (new_func, h)
    # func = new_func2
    # return new_func2
    return reduce(lambda f, g : lambda x:f(g(x)), func)
````

````python
def f(x):
    return x + 1

def g(x):
    return x * 2

def h(x):
    return x / 2
````

````python
new_func = compose2(f, g, h)
````

````python
new_func(10)
````

### 페아노의 공리

1. 1은 자연수이다.
2. 모든 자연수 a는 그 다음 자연수 b를 갖는다.
3. 1은 어떤 자연수의 다음 자연수가 아니다.
4. 두 자연수의 다음수가 같다면 두 자연수는 같다.
5. 어떤 수들의 집합이 1을 포함하고 모든 원소에 대해 그 다음수를 항상 포함한다면, 그 집합은 자연수이다.

## 페아노의 공리에 의한 자연수의 덧셈

`a + 0 = a`    
`a + b = 다음수(a) + 이전수(b)`

````python
def next_(n):
    return n + 1

def before(n):
    return n - 1

def add(n, a):
    if next_(a) == next_(1): # a와 1이 같다면.
        return next_(n)
    return add(next_(n), before(a)) # 재귀호출
````

````python
add(10, 20)
````

## 페아노의 공리에 의한 자연수의 곱셈

`a * 1 = a`     
`a * b = a + a * 이전수(b)`


````python
def mul(a, b):
    if next_(b) == next_(1):
        return a
    return add(a, mul(a, before(b)))
````

````python
mul(10, 10)
````

## 꼬리재귀

-> TCO 구현 X

````python
# for반복
def fib1(n):
    a = [1, 1]
    for _ in range(n-2):
        a.append(sum(a[-2:]))
    return a[-1]
````

````python
fib1(1000)
````

````python
# 재귀함수로 반복

def fib2(n):
    def increase(a, b, n):
        if n == 0:
            return a, b
        return increase(b, a+b, n-1)
    a, b = increase(0, 1, n)
    return a
````

````python
fib1(1000)
````

````python
fib2(1000)
````

````python
fib1(10000)
````

````python
fib2(10000)
````

````python
import random
````

````python
def make_list():
    return [ random.randint(1, 1000) for _ in range(100) ]
````

````python
def swap(x, i, j):
    x[i], x[j] = x[j], x[i]

def pivotFirst(x, lmark, rmark):
    pivot_val = x[lmark]
    pivot_idx = lmark
    while lmark <= rmark:
        while lmark <= rmark and x[lmark] <= pivot_val:
            lmark += 1
        while lmark <= rmark and x[rmark] >= pivot_val:
            rmark -= 1
        if lmark <= rmark:
            swap(x, lmark, rmark)
            lmark += 1
            rmark -= 1
    swap(x, pivot_idx, rmark)
    return rmark

def quickSort(x, pivotMethod=pivotFirst):
    def _qsort(x, first, last):
        if first < last:
            splitpoint = pivotMethod(x, first, last)
            _qsort(x, first, splitpoint-1)
            _qsort(x, splitpoint+1, last)
    _qsort(x, 0, len(x)-1)
    
    
target = make_list()
quickSort(target)
````

````python
def quick_sorted(arr):
    if len(arr) > 1:
        pivot = arr[len(arr)-1]
        left, mid, right = [], [], []
        for i in range(len(arr)-1):
            if arr[i] < pivot:
                left.append(arr[i])
            elif arr[i] > pivot:
                right.append(arr[i])
            else:
                mid.append(arr[i])
        mid.append(pivot)
        return quick_sorted(left) + mid + quick_sorted(right)
    else:
        return arr

target = make_list()
````

````python
def quick_sort(values):
    return quick_sort([x for x in values[1:] if x < values[0]]) + \
            [values[0]] + \
            quick_sort([x for x in values[1:] if x >= values[0]]) \
        if values else []
````

### curry

여러 인자를 받는 함수에서, 함수를 인자별로 세분화 하는것.

````python
import math


def fsum(f):
    def apply(a, b):
        return sum(map(f, range(a, b+1)))
    return apply
````

````python
log_sum = fsum(math.log)
````

````python
log_sum(1, 10)
````
