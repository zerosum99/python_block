## 리스트를 만드는 속도 비교

1. while 문 + append
2. for 문 + append
3. list comprehension
4. additional unpacking

```python
N = 100000
```

```python
%%time

result = []
i = 0
while True:
    result.append(i)
    i += 1
    if i > N:
        break
```

```python
%%time

result = []
for i in range(N):
    result.append(i)
```

```python
%%time

result = [ x for x in range(N) ]
```

```python
%%time

result = [ *range(N) ] #  == list(range(N))
```
