### 제너레이터 처리 

- def 와 yield 문장으로 제너레이터 객체를 만든다.


```python
def recursive_generator_(lis):
    if lis:
        yield lis[0]
        for i in recursive_generator_(lis[1:]) :
            yield i

for k in recursive_generator_([6,3,9,1]):
    print(k)
```

```python
def recursive_generator(lis):
    if lis:
        yield lis[0]
        yield from recursive_generator(lis[1:])

for k in recursive_generator([6,3,9,1]):
    print(k)
```

### 재귀호출에 대한 처리 제한 지정하기 

```python
import sys
sys.setrecursionlimit(3100)
```

```python
for k in recursive_generator([x for x in range(3000)]):
    print(k)
```

## 4.  사용자 정의 반복자 만들기 

```python
import collections.abc as cols
import abc


class Iterator(cols.Iterable):
    __slots__ = ()
    
    @abc.abstractmethod
    def __next__(self):
        "Return the next item from the iterator. When exhausted, raise StopIteration"
        raise StopIteration

    def __iter__(self):
        return self
    
    @classmethod
    def __subclasshook__(cls, C):
        if cls is Iterator:
            if (any("__next__" is B.__dict__ for B in C.__mro__) and
               any("__iter__" is B.__dict__ for B in C.__mro__)):
                return True
        return NotImplemented
```

### 반복형 문장 클래스를 만들기 

```python
import re
import reprlib

RE_WORD = re.compile('\w+')

class Sentence():
    def __init__(self, text):
        self.text = text
        self.words = RE_WORD.findall(text)
        
    def __getitem__(self, index):
        return self.words[index]
    
    def __len__(self):
        return len(self.words)
    
    def __repr__(self):
        return "Sentence(%s)" %reprlib.repr(self.text)
    
```

```python
print(dir(Sentence))
```

```python

s3 = Sentence("Pig and Pepper")
print(s3.words)
```

#### 반복자 만들기 

```python
import collections.abc as cols


class SeqIterator:
    def __init__(self, seq):
        self.seq = seq
        
    def __iter__(self):
        self.n = 0
        return self
    
    def __next__(self):
        if self.n < len(self.seq):
            result = self.seq[self.n]
            self.n += 1
            return result
        else:
            raise StopIteration
```

#### 반복자 생성해서 처리하기

```python
s = SeqIterator("반복자처리")
siter = iter(s)

print(siter)
print(issubclass(type(siter), cols.Iterator))
```

```python
for i in siter:
    print(i)
```

## 7. 반복자를 사용한 부분집합

```python
import itertools
def subset(n) :
    
    for i in range(0,n+1) :
        for i in itertools.combinations(range(0,n),i) :
            print(i)
            subset.s.append(set(i))
```

```python
subset.s = []
```

```python
subset(3)
```

```python
subset.s
```

### 순환문을 이용한 부분집합

```python
import copy 

def subset2(args) :
    
    args = list(set(args))
    print(args)
    powerSetList = [set()]
    for i in range(len(args)) :
        count = len(powerSetList)
        print(count)
        for j in range(0,count) :
            ## 첫번째 부분 집합은 항상 공집합
            tempSet = copy.deepcopy(powerSetList[j])
            ### 리스트의 순서적인 원소를 기준으로  생성
            tempSet.add(args[i])
            print(tempSet)
            powerSetList.append(tempSet)
    return powerSetList
```

```python
subset2([1,1,1,2,3])
```
