```python
for i in  df.columns :
    if df[i].dtype == np.int64 :
        df[i] = df[i].astype(np.int16)
    if df[i].dtype == np.float64 :
        df[i] = df[i].astype(np.float32)
```

```python
d_ = df [['성별코드','신장(5Cm단위)','연령대코드(5세단위)']].copy()
```

```python
d_.shape
```

```python
d_["count"]  = 1
```

```python
d_x = pd.pivot_table(d_, values='count', index=['신장(5Cm단위)'],columns=['성별코드','연령대코드(5세단위)']
                    , aggfunc=np.sum).fillna(0)
```
